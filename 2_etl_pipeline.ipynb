```python
import pandas as pd
import numpy as np

print("--- STEP 1: EXTRACTION PHASE ---")

df_web = pd.read_json('web_data.json', lines=True)
df_crm = pd.read_excel('crm_data.xlsx')
df_finance = pd.read_csv('finance_data.csv')

print(f"Web Data: {df_web.shape}")
print(f"CRM Data: {df_crm.shape}")
print(f"Finance Data: {df_finance.shape}")

print("\n--- STEP 2: TRANSFORMATION PHASE (Merging) ---")

df_merged = pd.merge(df_crm, df_web, on='reservation_id')
df_final = pd.merge(df_merged, df_finance, on='reservation_id')

print(f"Merged Data Warehouse Shape: {df_final.shape}")

print("\n--- STEP 3: DATA CLEANING ---")

df_final['reservation_status_date'] = pd.to_datetime(df_final['reservation_status_date'], dayfirst=False, errors='coerce')

df_final['adr'] = df_final.groupby('hotel')['adr'].transform(lambda x: x.fillna(x.median()))

df_final['country'] = df_final['country'].fillna('Unknown')

df_final.dropna(subset=['reservation_status_date'], inplace=True)

print("Cleaning completed successfully.")

print("\n--- STEP 4: LOADING PHASE ---")

df_final.to_csv('data_warehouse_cleaned.csv', index=False)

print("SUCCESS: 'data_warehouse_cleaned.csv' has been created!")
```
