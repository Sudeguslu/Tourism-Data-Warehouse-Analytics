```python
import pandas as pd
import numpy as np

print("Loading original data...")
df = pd.read_csv('hotel_bookings.csv')

df.insert(0, 'reservation_id', range(1, len(df) + 1))

web_columns =['reservation_id', 'country', 'meal', 'required_car_parking_spaces', 'total_of_special_requests']
df_web = df[web_columns].copy()
df_web.to_json('web_data.json', orient='records', lines=True)
print("Web data saved as 'web_data.json'.")

crm_columns =['reservation_id', 'hotel', 'is_canceled', 'lead_time', 'arrival_date_year', 
               'arrival_date_month', 'arrival_date_week_number', 'arrival_date_day_of_month', 
               'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies']
df_crm = df[crm_columns].copy()
df_crm.to_excel('crm_data.xlsx', index=False)
print("CRM data saved as 'crm_data.xlsx'.")

finance_columns =['reservation_id', 'market_segment', 'distribution_channel', 'is_repeated_guest', 
                   'reserved_room_type', 'assigned_room_type', 'deposit_type', 'agent', 'company', 
                   'customer_type', 'adr', 'reservation_status', 'reservation_status_date']
df_finance = df[finance_columns].copy()

np.random.seed(42)
random_null_indices = np.random.choice(df_finance.index, size=5000, replace=False)
df_finance.loc[random_null_indices, 'adr'] = np.nan

def mess_up_dates(date_str):
    if pd.isna(date_str):
        return date_str
    if np.random.rand() > 0.5:
        parts = str(date_str).split('-')
        if len(parts) == 3:
            return f"{parts[2]}/{parts[1]}/{parts[0]}"
    return date_str

df_finance['reservation_status_date'] = df_finance['reservation_status_date'].apply(mess_up_dates)

df_finance.to_csv('finance_data.csv', index=False)
print("Finance data purposely messed up and saved as 'finance_data.csv'.")

print("\nData chaos successfully created!")
```
